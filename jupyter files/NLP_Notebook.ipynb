### Loading text corpora and pre-processing

Load the text corpus `state_union` from the `corpus` module. It contains the "state of the union" speeches of the US presidents.

Further information at: https://www.nltk.org/api/nltk.corpus.html#module-nltk.corpus

```python
from nltk.corpus import state_union
```

Use the `fileids` method to display the names of the text data files in `state_union` 

```python
state_union.fileids()
```

Load with the `raw` method the speech "2001-GWBush-2.txt" and store it in the variable `bush2001`

```python
# .raw() is a method provided by NLTK. The method is used to access the raw text content of a document in the corpus.

bush2001 = state_union.raw("2001-GWBush-2.txt")
```

```python
type(bush2001)
```

```python
# bush2001 is a string
bush2001
```

How many characters are in `bush2001`?

```python
len(bush2001)
```

Count the number of sentences

```python
from nltk.tokenize import sent_tokenize
```

```python
bush_sentences = sent_tokenize(bush2001)
```

```python
bush_sentences
```

```python
# let us find out what kind of variable bush_sentence is
type(bush_sentences)
```

```python
# thus, we deal with a list
# let us print out the first element to check what this list is
print([bush_sentences[0]])
print([bush_sentences[1]])
```

```python
# obviously, each element in the list is a sentence. Thus, we can count the number of sentences by using the length function again
len(bush_sentences)
```

Use `word_tokenize` to split `bush2001` into words and store your result as `bush2001_word`

```python
from nltk.tokenize import word_tokenize

bush2001_word = word_tokenize(bush2001)
```

```python
bush2001_word
```

Number of words

```python
# again, we created a list, where each element is a word
len(bush2001_word)
```

Change the text into lower-case letters 

```python
bush2001_lower = [t.lower() for t in bush2001_word]
bush2001_lower
```

Filter out numbers/ symbols. Use the `isalpha` method

```python
bush2001_alpha = [t for t in bush2001_lower if t.isalpha()]
#isalpha() method returns true if all characters in the string are alphabetic and there is at least one character and false otherwise.
bush2001_alpha
```

### Frequency distribution of words

```python
from nltk.probability import FreqDist
```

Create a `FreqDist` object for `bush2001_alpha` 

```python
bush2001_fdist = FreqDist(bush2001_alpha)
bush2001_fdist
```

How many unique words are in `bush2001_fdist` ?

```python
len(bush2001_fdist)
```

Print the 40 most common words with the `most_common` method

```python
bush2001_fdist.most_common(40)
```

Load the module `stopwords` from `corpus` 

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

Store the English stop words in the variable `stop_words`
https://www.nltk.org/howto/corpus.html

```python
stop_words = stopwords.words("english")
```

```python
stop_words
```

Write a `for`-loop on `bush2001_alpha`, which erases each word that is in `stop_words`. Store your results as `bush2001_filtered`.

```python
# bush2001_alpha= speech in lower letters without symbols and numbers

bush2001_filtered = []
for w in bush2001_alpha:
    if w not in stop_words:
        bush2001_filtered.append(w)
```

```python
# unsurprisingly, we created a new list
bush2001_filtered
```

Alternative:
bush2001_filtered = [w for w in bush2001_alpha if w not in stop_words]

Create a frequency distribution for `bush2001_filtered` and store it as `bush2001_filtered_fdist`.

```python
bush2001_filtered_fdist = FreqDist(bush2001_filtered)
bush2001_filtered_fdist
```

#### Plotting a frequency distribution


```python
# Frequency Distribution Plot
import matplotlib.pyplot as plt
bush2001_filtered_fdist.plot(20, cumulative = False)
plt.show()
```

### N-grams



Instead of looking at single worlds, we now look at a set of co-occuring words within a given window

```python
from nltk.util import ngrams
bush2001_bigrams = ngrams(bush2001_filtered,2)
```

```python
bush2001_bigrams_fdist = FreqDist(bush2001_bigrams)
print(bush2001_bigrams_fdist)
```

```python
bush2001_bigrams_fdist
```

```python
len(bush2001_bigrams_fdist)
```

```python
bush2001_bigrams_fdist.most_common(30)
```

```python
bush2001_bigrams_fdist.plot(20, cumulative = False)
plt.show()
```

```python
# lets play around
bush2001_3grams = ngrams(bush2001_filtered,3)
bush2001_3grams_fdist = FreqDist(bush2001_3grams)
print(bush2001_3grams_fdist)
```

```python
bush2001_3grams_fdist
```

```python
bush2001_3grams_fdist.most_common(30)
```

```python
bush2001_3grams_fdist.plot(20, cumulative = False)
plt.show()
```

## Word clouds

Word clouds display the frequencies of words in a text corpus: the more frequent the word, the bigger will be the word in the word cloud 

```python
!pip install wordcloud
```

```python
from wordcloud import WordCloud
```

### Speech from G.W. Bush

```python
bush2001_filtered
```

```python
type(bush2001_filtered)
```

```python
bush_str=' '.join(bush2001_filtered)
bush_str
```

```python
type(bush_str)
```

```python
bush2001_wordcloud = WordCloud().generate(bush_str)
```

If do not set any parameters in WordCloud(), then the default values will be taken

```python
plt.figure(figsize=(40, 30)) # Set figure size
plt.axis("off") # No axis details
plt.imshow(bush2001_wordcloud); # Display image
```

Using the original data set from the Bush speech 2001, containing no pre-processing steps
